We'll continue to make use of the fuel economy dataset in this workspace.

```python
fuel_econ = pd.read_csv('./data/fuel_econ.csv')
fuel_econ.head()
```

**Task**: What is the relationship between the size of a car and the size of its engine? The cars in this dataset are categorized into one of five different vehicle classes based on size. Starting from the smallest, they are: {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, and Large Cars}. The vehicle classes can be found in the 'VClass' variable, while the engine sizes are in the 'displ' column (in liters). **Hint**: Make sure that the order of vehicle classes makes sense in your plot!

```python
car_class = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# r
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=car_class)

# Use pandas.astype() to convert the "VClass" column from a plain object type into an ordered categorical type 
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)

# Resize figure to accommodate two plots

#plt.rcParams['figure.figsize'] = [20, 5]

plt.figure(figsize = [20, 5])

plt.subplot(1,2,1)
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'displ', color = sb.color_palette()[0])
plt.xticks(rotation=40);

plt.subplot(1,2,2)
sb.boxplot(data = fuel_econ, x = 'VClass', y = 'displ', color = sb.color_palette()[0])

interpretation = "The relationship between the variables is not consistent. Compact car has smaller engine when compared to cars relatively smaller than it (Minicompact and SUbcompact)."
print(interpretation)
```
